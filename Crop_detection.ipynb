```python


from google.colab import files
from PIL import Image
import numpy as np

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  try:
    img = Image.open(fn)
    img = img.convert("RGB")

    cropped_img = img.crop((50, 50, 200, 200))

    img_array = np.array(cropped_img)

    pixel_x = 10
    pixel_y = 10
    r, g, b = img_array[pixel_y, pixel_x]
    print(f"RGB values at ({pixel_x}, {pixel_y}): R={r}, G={g}, B={b}")

    average_r = np.mean(img_array[:,:,0])
    average_g = np.mean(img_array[:,:,1])
    average_b = np.mean(img_array[:,:,2])
    print(f"Average RGB values: R={average_r:.2f}, G={average_g:.2f}, B={average_b:.2f}")


  except Exception as e:
    print(f"Error processing image: {e}")
```

```python


from google.colab import files
from PIL import Image
import numpy as np

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  try:
    img = Image.open(fn)
    img = img.convert("RGB")

    img_array = np.array(img)

    region_x1, region_y1, region_x2, region_y2 = 50, 50, 200, 200
    region = img_array[region_y1:region_y2, region_x1:region_x2, 0]
    hottest_pixel_value = np.max(region)
    hottest_pixel_coordinates = np.unravel_index(np.argmax(region), region.shape)

    hottest_pixel_x = hottest_pixel_coordinates[1] + region_x1
    hottest_pixel_y = hottest_pixel_coordinates[0] + region_y1

    print(f"Hottest pixel in region ({region_x1},{region_y1})-({region_x2},{region_y2}):")
    print(f"  Value: {hottest_pixel_value}")
    print(f"  Coordinates: ({hottest_pixel_x}, {hottest_pixel_y})")

    average_temperature = np.mean(region)
    print(f"Average temperature in region: {average_temperature:.2f}")


  except Exception as e:
    print(f"Error processing image: {e}")
```

```python

from google.colab import files
from PIL import Image
import numpy as np
import random

def analyze_image(image_path):
    try:
        img = Image.open(image_path)
        img = img.convert("RGB")
        img_array = np.array(img)

        pixel_x, pixel_y = 10, 10
        r, g, b = img_array[pixel_y, pixel_x]
        print(f"RGB values at ({pixel_x}, {pixel_y}): R={r}, G={g}, B={b}")

        average_r = np.mean(img_array[:,:,0])
        average_g = np.mean(img_array[:,:,1])
        average_b = np.mean(img_array[:,:,2])
        print(f"Average RGB values: R={average_r:.2f}, G={average_g:.2f}, B={average_b:.2f}")


        region_x1, region_y1, region_x2, region_y2 = 50, 50, 200, 200
        region = img_array[region_y1:region_y2, region_x1:region_x2, 0]
        hottest_pixel_value = np.max(region)
        hottest_pixel_coordinates = np.unravel_index(np.argmax(region), region.shape)

        hottest_pixel_x = hottest_pixel_coordinates[1] + region_x1
        hottest_pixel_y = hottest_pixel_coordinates[0] + region_y1

        print(f"Hottest pixel in region ({region_x1},{region_y1})-({region_x2},{region_y2}):")
        print(f"  Value: {hottest_pixel_value}")
        print(f"  Coordinates: ({hottest_pixel_x}, {hottest_pixel_y})")

        average_temperature = np.mean(region)
        print(f"Average temperature in region: {average_temperature:.2f}")

        simulated_depth_data = [random.randint(50, 200) for _ in range(10)]
        print("Simulated depth data (distances in cm):", simulated_depth_data)

    except Exception as e:
        print(f"Error processing image: {e}")


uploaded = files.upload()
for fn in uploaded.keys():
    print(f'User uploaded file "{fn}"')
    analyze_image(fn)
```

**Checking the crop whether is health or not**

```python

def analyze_crop_health(image_path):
    try:
        img = Image.open(image_path)
        img = img.convert("RGB")
        img_array = np.array(img)

        average_r = np.mean(img_array[:,:,0])
        average_g = np.mean(img_array[:,:,1])
        average_b = np.mean(img_array[:,:,2])

        r_threshold = 150
        g_threshold = 150
        b_threshold = 150

        if average_r > r_threshold and average_g > g_threshold and average_b > b_threshold:
          print("Crop health: Good (Healthy color balance)")
        else:
          print("Crop health: Not Good (Unhealthy color balance)")

    except Exception as e:
        print(f"Error analyzing crop health: {e}")


uploaded = files.upload()
for fn in uploaded.keys():
  print(f'User uploaded file "{fn}"')
  analyze_crop_health(fn)
```
